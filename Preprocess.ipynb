```python
import re
```

```python
from pymystem3 import Mystem
import pymorphy2
```

```python
m = Mystem()
```

```python
f= open(r'zhivi-i-pomni-5.txt','r',encoding='utf-8')
```

```python
def cut_sentences(content):
    sentences=re.split(r'(\.|\!|\?|\.{3})',content)
    return sentences
```

```python
content=''.join(f.readlines())
```

```python
sentences=cut_sentences(content)
print('\n'.join(sentences))
```

```python
import string 
s = '\n'.join(sentences) 
for c in string.punctuation: 
    s = s.replace(c,'') 
print(s)
print(s.strip())
```

```python
lemmas = m.lemmatize(s) 
lemmas[10:20]
```

```python
fh=open('zhivi-i-pomni.txt', 'w', encoding='utf-8')
fh.write(''.join(lemmas))
fh.close()
```
